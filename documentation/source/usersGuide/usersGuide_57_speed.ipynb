## Running processes in parallel on multiple cores.

```python
def countNotesOneWork(metadataObject):
    # print(metadataObject)
    work = metadataObject.parse()
    return len(work.recurse().notes)
```

```python
allLudwig = corpus.search('beethoven')
allLudwig
```

```python
results = common.runParallel(allLudwig, countNotesOneWork)
```

```python
results
```

```python
def runNormal(allLudwig):
    results = []
    for metadataObject in allLudwig:
        result = countNotesOneWork(metadataObject)
        results.append(result)
    return results
```

```python
%timeit common.runParallel(allLudwig, countNotesOneWork)
```

```python
%timeit runNormal(allLudwig)
```
